```python
import pandas as pd

# Fake sales data
data = {
    'Product': ['Protein Powder', 'Dumbbells', 'Yoga Mat', 'Shaker'],
    'Price': [1200, 2500, 800, 400],
    'Units_Sold': [10, 5, 2, 15]
}

df = pd.DataFrame(data)

# Show total revenue for each product
df['Revenue'] = df['Price'] * df['Units_Sold']
print("Full Table:\n", df)

# Show which product made the most money
top = df.loc[df['Revenue'].idxmax()]
print("\nBest Seller:\n", top)
```
